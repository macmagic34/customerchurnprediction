```python
## using stat tools to check which features are suitable for training

from scipy.stats import chi2_contingency, ttest_rel
from sklearn.tree import ExtraTreeClassifier
from sklearn.feature_selection import SelectKBest, f_classif
```

```python
sk = SelectKBest(f_classif, k=4)
sk.fit(X_train, y_train)

len(sk.scores_)
```

```python
sk.scores_
```

```python
sk.get_support(indices=True)
```

```python
score_df = pd.DataFrame({'Features' : df4.columns[0:-1], 'Scores' : sk.scores_})
```

```python
score_df.nlargest(10, columns='Scores')
```

```python
best_features = list(score_df.nlargest(10, columns='Scores')['Features'])
best_features
```

```python
X_train = X_train[best_features]
X_test = X_test[best_features]
print(X_train.shape)
print(X_test.shape)
```

```python
## now we will scale the data since we might implement logistic regression ahead

from sklearn.preprocessing import StandardScaler

ss = StandardScaler()
scaled_features = ss.fit_transform(X_train[['tenure', 'MonthlyCharges']])
X_train.reset_index(drop=True, inplace=True)
X_train.drop(['tenure', 'MonthlyCharges'], inplace=True, axis=1)
scaled_features = pd.DataFrame(scaled_features, columns=['tenure', 'MonthlyCharges'])
X_train = pd.concat([X_train, scaled_features], axis=1)
X_train.head()
```

```python
X_train.shape
```

```python
scaled_features = ss.transform(X_test[['tenure', 'MonthlyCharges']])
X_test.reset_index(drop=True, inplace=True)
X_test.drop(['tenure', 'MonthlyCharges'], inplace=True, axis=1)
scaled_features = pd.DataFrame(scaled_features, columns=['tenure', 'MonthlyCharges'])
X_test = pd.concat([X_test, scaled_features], axis=1)

X_test.head()
X_test = X_test.astype('float')
```

```python
X_test.shape
```

```python
X_train.shape
```

```python
## Now our data is cooked and is scaled to our requirement. let us start with the model building now

from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV, KFold
```

```python
values = {
    'logistic_regression' : {
        'model' : LogisticRegression(),
        'params' : {
            'C' : [0.001, 0.01, 0.1, 1, 10, 100],
            'solver' : ['newton-cg', 'lbfgs', 'liblinear', 'sag', 'saga']
        }
    },
    
    'random_forest_classifier' : {
        'model' : RandomForestClassifier(),
        'params' : {
            'n_estimators' : [1, 10, 100, 1000]
        }
    },
    
    'svc' : {
        'model' : SVC(),
        'params' : {
            'C' : [0.001, 0.01, 0.1, 1, 10, 100],
            'kernel' : ['linear', 'poly', 'rbf']
        }
    },
}
```

```python
scores=[]
for model_name, model in values.items():
    gscv = GridSearchCV(estimator=model['model'], param_grid=model['params'], return_train_score=False, cv=5)
    gscv.fit(X_train, y_train)
    scores.append({
        'model' : model_name,
        'best_score' : gscv.best_score_,
        'best_params' : gscv.best_params_
    })
model_df = pd.DataFrame(scores, columns=['model', 'best_score', 'best_params'])
```

```python
model_df
```

```python
lr = LogisticRegression(solver='newton-cg', C=0.1)
lr.fit(X_train, y_train)
lr.score(X_test, y_test)
```

```python
from sklearn.metrics import classification_report, roc_curve, plot_confusion_matrix
```

```python
y_pred = lr.predict(X_test)
cr = classification_report(y_test, y_pred)
print(cr)
```

```python
fpr, tpr, _ = roc_curve(y_test, y_pred)
plt.plot(fpr, tpr)
```

```python
plot_confusion_matrix(lr, X_train, y_train)
```
